```python
def custom_mean(data):
    sum(data)/len(data)

def custom_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        return (sorted_data[(n//2)-1] + sorted_data[n//2])/2
    else:
        return sorted_data[n//2]

def custom_var(data):
    mean = custom_mean(data)
    n = len(data)
    return (sum((x - mean) ** 2 for x in data)) / (n)

def custom_std(data):
    return custom_var ** 0.5

def custom_range(data):
    return max(data) - min(data)

def custom_percentile(data, percent):
    sorted_data = sorted(data)
    k = (len(data) - 1) * percent / 100
    f = int(k)
    c = k - f
    if f+1 < len(sorted_data)-1:
        return sorted_data[f] + (sorted_data[f+1] - sorted_data[f])*c
    else:
        return sorted_data[f]
```
