```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
df.isnull().sum()
```

```python
 df.drop('deck', axis=1, inplace= True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='age', y='sex', hue='survived')
plt.title('Boxplot by sex and age distribution')
plt.show()
# ObservationA higher proportion of females survived compared to males, as seen by the larger orange section (survived = 1).
# Female survivors generally ranged from 15 to 40 years, with some even younger.
# Female non-survivors had a slightly broader age spread but fewer in number.
# Most males did not survive, indicated by a larger blue section (survived = 0).
# Male survivors were fewer and mostly aged between 20 and 35.
# Some outliers in male age (above 60) are visible, especially among non-survivors.
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='sex', y='age', hue='survived')
plt.title("Violin plot of Age distribution by gender and survival")
plt.show()
#This plot shows that the number of female survivor is more as compared to male and the age of the most survived female lies between 15-35 and the number of most 
#survived male lies between 20 -35 and the number of males having age greater than 60 and survived is less 
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='sex', y='age', hue='survived')
plt.title("Swarmplot of Age distribution by gender and survival")
plt.show()
#The swramplot tells that the number of survived female is more as compared to male and the number of male is more having age between 15-75 there are also
#some male child whose age is between 0-2 and there surived number is more 
#Dense clusters show many young female survivors.
#Male survivors are fewer and scattered.
```
