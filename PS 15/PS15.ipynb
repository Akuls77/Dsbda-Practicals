```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import re
import nltk
import string
```

```python
df = pd.read_csv("Alexa-Dataset - Alexa-Dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
sns.countplot(data = df, x='feedback', palette='Set2')
plt.xlabel('Feedback(0 = Negative and 1 = Positive)')
plt.ylabel('Count')
plt.show()
```

```python
df['lower_text'] = df['verified_reviews'].str.lower()
```

```python
df
```

```python
def remove_punctuation(data):
    puncfree = "".join([i for i in data if i not in string.punctuation])
    return puncfree

df['remove_punc'] = df['lower_text'].apply(lambda x: remove_punctuation(x))
```

```python
df
```

```python
import emoji
```

```python
def remove_emoji(text):
    if isinstance(text, str):
        return emoji.replace_emoji(text, replace="")
    return text

df['remove_emoji'] = df['remove_punc'].apply(lambda x: remove_emoji(x))
```

```python
df
```

```python
def tokenization(text):
    token = re.split(r'\W+', text)
    return token

df['tokenized_string'] = df['remove_emoji'].apply(lambda x: tokenization(x))
```

```python
df
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
",".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df['remove_stopword'] = df['remove_emoji'].apply(lambda x: remove_stopwords(x))
```
