```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bp_df = pd.read_csv("Banglore_Housing_Prices.csv")
bp_df
```

```python
bp_df.isnull().sum()
```

```python
bp_df.dropna(inplace =True)
```

```python
bp_df.isnull().sum()
```

```python
bp_df.info()
```

```python
bp_df['size'] = bp_df['size'].str.split().str[0].astype(int)
bp_df
```

```python
def convert_tsqft(x):
    tokens = x.split('-')
    if(len(tokens)==2):
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
bp_df['total_sqft'] = bp_df['total_sqft'].apply(convert_tsqft)
```

```python
bp_df
```

```python
bp_df["price_per_sqft"] = bp_df['price'] / bp_df['total_sqft']
bp_df
```

```python
bp_df.isnull().sum()
```

```python
bp_df.dropna(inplace = True)
```

```python
num_col = ['size','price_per_sqft']
```

```python
sns.boxplot(data = bp_df[num_col])
plt.show()
```

```python
def detect_outliers(data):
    outlier = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lb = q1 - (1.5*iqr)
    ub = q3 + (1.5*iqr)
    for i in data:
        if (i<lb or i>ub):
            outlier.append(i)
    return outlier, lb, ub
```

```python
detect_outliers(bp_df['price_per_sqft'])
```

```python
data = sorted(bp_df['price_per_sqft'])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lb = q1 - (1.5*iqr)
ub = q3 + (1.5*iqr)

bp_df['price_per_sqft'] = bp_df['price_per_sqft'].apply(lambda x: lb if x<=lb else ub if x>=ub else x)
```

```python
detect_outliers(bp_df['size'])
```

```python
data = sorted(bp_df['size'])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lb = q1 - (1.5*iqr)
ub = q3 + (1.5*iqr)

bp_df['size'] = bp_df['size'].apply(lambda x: lb if x<=lb else ub if x>=ub else x)
```

```python
sns.boxplot(data = bp_df[num_col])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

X = bp_df[['size', 'total_sqft', 'bath']]
y = bp_df['price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Performance Metrics
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
```

```python
print("Training MSE:", train_mse)
print("Testing MSE:", test_mse)
print("Training R² (Accuracy):", train_r2)
print("Testing R² (Accuracy):", test_r2)
```
