```python
nba_df['Salary'] = nba_df['Salary'].fillna(nba_df['Salary'].mean())
```

```python
nba_df['College'] = nba_df['Salary'].fillna(nba_df['College'].mode())
```

```python
nba_df.isnull().sum()
```

```python
nba_df['Salary'] = nba_df['Salary'].astype('int')
```

```python
nba_df.dtypes
```

```python
nba_df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
nba_df[['Age','Weight','Salary']] = scaler.fit_transform(nba_df[['Age','Weight','Salary']])
```

```python
nba_df
```

```python
nba_df['Position'].unique()
```

```python
nba_df['Position'].replace(['PG', 'SF', 'SG', 'PF', 'C'],[0,1,2,3,4])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
nba_df['Team'] = le.fit_transform(nba_df['Team'])
```
