```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import string
```

```python
df = pd.read_csv("Alexa-Dataset - Alexa-Dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['lower_text'] = df['verified_reviews'].str.lower()
```

```python
df
```

```python
def remove_punc(text):
    puncfree = "".join([word for word in text if word not in string.punctuation])
    return puncfree

df['remove_punctuation'] = df['lower_text'].apply(lambda x: remove_punc(x))
```

```python
df
```

```python
def tokenization(text):
    token = re.split(r'\W+', text)
    return token

df['tokenized_string'] = df['remove_punctuation'].apply(lambda x: tokenization(x))
```

```python
df
```

```python
from nltk.corpus import stopwords
```

```python
# nltk.download('stopwords')
stopwords.words('english')
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
```

```python
df['rem_stopword'] = df['remove_punctuation'].apply(lambda x: remove_stopwords(x))
df
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def stemming(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
```

```python
df['stemmed_text'] = df['remove_punctuation'].apply(lambda x: stemming(x))
df
```

```python
# nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatization(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['lemmatized_text'] = df['remove_punctuation'].apply(lambda x: lemmatization(x))
df
```
