```python
def detect_outlier(data):
    data = sorted(data)
    outlier = []
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lb = q1 - 1.5*iqr
    ub = q3 + 1.5*iqr
    for i in data:
        if (i<lb or i>ub):
            outlier.append(i)
    return outlier, lb, ub
```

```python
detect_outlier(df['sepal_width'])
```

```python
sns.boxplot(data = df, x='sepal_width')
```

```python
q1 = np.percentile(df['sepal_width'], 25)
q3 = np.percentile(df['sepal_width'], 75)
iqr = q3 - q1
lb = q1 - 1.5*iqr
ub = q3 + 1.5*iqr

df['sepal_width'] = df['sepal_width'].apply(lambda x: lb if x<=lb else ub if x>=ub else x)
```

```python
detect_outlier(df['sepal_width'])
```

```python
sns.boxplot(data = df, x='sepal_width')
```
