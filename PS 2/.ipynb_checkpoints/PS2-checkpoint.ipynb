```python
def get_mean(data):
    return sum(data)/len(data)
```

```python
def get_median(data):
    sorted_data = sorted(data)
    n=len(sorted_data)
    if n%2 == 0:
        return (sorted_data[(n//2)-1] + sorted_data[n//2])/2
    else:
        return (sorted_data[n//2])
```

```python
def get_min(data):
    min_val = data[0]
    for val in data:
        if val < min_val:
            min_val = val
    return min_val
```

```python
def get_max(data):
    max_val = data[0]
    for val in data:
        if val > max_val:
            max_val = val
    return max_val
```

```python
def get_std(data):
    mean = get_mean(data)
    variance = sum((x-mean)**2 for x in data) / (len(data))
    return variance**0.5
```

```python
print("Mean:",get_mean(df['Income']))
print("Median:",get_median(df['Income']))
print("Min:",get_min(df['Income']))
print("Max:",get_max(df['Income']))
print("Std:",get_std(df['Income']))
```

```python
np_array = np.array(df['Income'])
pd_array = pd.Series(df['Income'])
```

```python
# Numpy method

from scipy import stats
np_mean = np.mean(np_array)
print(np_mean)
np_median = np.median(np_array)
print(np_median)
np_mode = stats.mode(np_array)
print(np_mode)

np_var = np.var(np_array, ddof=1)
print(np_var)
np_std = np.std(np_array, ddof=1)
print(np_std)
```

```python
# Pandas Method
pd_mean = pd_array.mean()
print(pd_mean)
pd_median = pd_array.median()
print(pd_median)
pd_mode = pd_array.mode()
print(pd_mode)

pd_var = pd_array.var(ddof=1)
print(pd_var)
pd_std = pd_array.std(ddof=1)
print(pd_std)
```

```python
# Statistics Method

import statistics as st
s_mean = st.mean(df['Income'])
print(s_mean)
s_median = st.median(df['Income'])
print(s_median)
s_mode = st.mode(df['Income'])
print(s_mode)
```
